## Literacy Status Data Analysis and Visualization of Gandaki Province Nepal

In this dataset form NPHC 2021, we have literacy status data of every district of Gandaki province.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset:

```python
df = pd.read_csv("./Data/LiteracyStatus.csv")
```

Creating a dataframe only with district's data:


```python
district_df = df.drop([0,1,2,3,4,5])
```

```python
district_df
```

```python
district_total = district_df[district_df['Gender'] == 'Total']
```

```python
district_total
```

```python
district_total = district_total.drop(["Gender"], axis = 1)
```

```python
district_total
```

```python
area_df = district_total.groupby(by='Area').sum()
area_df
```

Pie chart distribution of literacy status of each district:

```python
area_df['Total_Population'].plot(kind = 'pie', figsize=(20,10), shadow=True, autopct='%1.1f%%')
plt.title(' Population distribution by Area')
plt.axis('equal')
plt.show()
```

```python
area_df['Can_Read_Write'].plot(kind = 'pie', figsize=(20,10), shadow=True, autopct='%1.1f%%')
plt.title(' Population distribution by Area')
plt.axis('equal')
plt.show()
```

Bar chart distribution of literacy status of each district:

```python
area_df.plot(kind='bar', color=['skyblue', 'red','green','blue','orange'])
plt.xlabel('Area')
plt.ylabel('Total_Population')
plt.title('Population according to literacy status')
plt.show()
```

```python
area_df.plot(kind='bar', color=['skyblue', 'red','green','blue','orange'])
plt.xlabel('Area')
plt.ylabel('Can_Read_Write')
plt.title('Population according to literacy status')
plt.show()
```

```python
df_new = district_total.copy()
```

```python
df_new.head()
```

```python
df_new.set_index('Area',inplace=True)
```

```python
df_new.sort_values(by='Total_Population', ascending=True, inplace=True)

df_top10 = df_new['Total_Population'].tail(5)
df_top10
```

```python
df_top10.plot(kind='barh', figsize=(10, 10), color='darkblue',)
plt.xlabel('Population')
plt.title('Top 5 Districts With High Population')

plt.show()
```

```python
df_new.set_index('Area',inplace=True)

df_new.sort_values(by='Can_Read_Write', ascending=True, inplace=True)

df_top10 = df_new['Can_Read_Write'].tail(5)
df_top10

df_top10.plot(kind='barh', figsize=(10, 10), color='darkblue',)
plt.xlabel('Population')
plt.title('Top 5 Districts Population Who Can Read and Write')

plt.show()
```

```python
df_new.set_index('Area',inplace=True)

df_new.sort_values(by='Can_Read_only', ascending=True, inplace=True)

df_top10 = df_new['Can_Read_only'].tail(5)
df_top10

df_top10.plot(kind='barh', figsize=(10, 10), color='darkblue',)
plt.xlabel('Population')
plt.title('Top 5 Districts Population Who Can Read Only')

plt.show()
```

```python
df_new.set_index('Area',inplace=True)

df_new.sort_values(by="Can't_Read_Write", ascending=True, inplace=True)

df_top10 = df_new["Can't_Read_Write"].tail(5)
df_top10

df_top10.plot(kind='barh', figsize=(10, 10), color='green',)
plt.xlabel('Population')
plt.title('Top 5 Districts Population Who Cant Read and Write')

plt.show()
```
